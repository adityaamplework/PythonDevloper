### 4.symantic Splitting with Embedding

```python
text=text[0].page_content
```

```python
import numpy as np
from langchain_huggingface import HuggingFaceEmbeddings

embedding_model = HuggingFaceEmbeddings(model_name='all-MiniLM-L6-v2')

sentences = text.split(". ")

sentence_embeddings = embedding_model.embed_documents(sentences)


def cosine_similarity(vec1, vec2):
    return np.dot(vec1, vec2) / (np.linalg.norm(vec1) * np.linalg.norm(vec2))


THRESHOLD = 0.8
split_indices = []
for i in range(len(sentence_embeddings) - 1):
    if cosine_similarity(sentence_embeddings[i], sentence_embeddings[i + 1]) < THRESHOLD:
        split_indices.append(i)

chunks = []
start_idx = 0
for idx in split_indices:
    chunks.append(". ".join(sentences[start_idx:idx + 1]) + ".")
    start_idx = idx + 1
chunks.append(". ".join(sentences[start_idx:]) + ".")

print(chunks)
```
