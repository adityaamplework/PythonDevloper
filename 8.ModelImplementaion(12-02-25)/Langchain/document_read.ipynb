Pdf loader

```python
from langchain_community.document_loaders import PyPDFLoader
pdf=PyPDFLoader("attention.pdf")
docs=pdf.load()
```

Web Loader

```python
from langchain_community.document_loaders import WebBaseLoader
loader=WebBaseLoader(web_path='https://en.wikipedia.org/wiki/India')
loader.load()
```

```python
from langchain_community.document_loaders import ArxivLoader

loader=ArxivLoader(query='1706.03762 ',doc_content_chars_max=2)
```

```python
loader.load()
```

Wikipidea Loader

```python
from langchain_community.document_loaders import WikipediaLoader
loader=WikipediaLoader(query='Transformers in  ml',load_max_docs=2)
loader.load()
```

Text_spilleter


```python
from langchain_text_splitters  import RecursiveCharacterTextSplitter
text_split=RecursiveCharacterTextSplitter(chunk_size=500,chunk_overlap=50)
final_split=text_split.split_documents(docs)
```
